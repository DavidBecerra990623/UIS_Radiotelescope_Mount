```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Star = "Sadalmelik"
star_df = pd.read_csv("16_Sadalmelik.csv",sep = ',')
star_df.columns
```

```python
Hora = np.array(star_df["Hora"])
azimut = np.array(star_df["Azimut"])
elev = np.array(star_df["Elevaci贸n"])
vel_az = np.array(star_df["Velocidad Azimut"])
vel_el = np.array(star_df["Velocidad Elevaci贸n"])
```

```python
#Grafica de posicion

xpos_stick = len(Hora)
fig, ax = plt.subplots(figsize=(10,6))
azimut_plot = ax.plot(Hora, azimut ,label="Azimut")
elevacion_plot = ax.plot(Hora, elev,label="Elevaci贸n")
ax.set_ylabel("Azimut - Elevacion")
ax.set_xlabel("Tiempo")
xpos_values = ax.get_xticks()
newx_pos_values = [xpos_values[i] for i in range(0,xpos_stick,xpos_stick//10)]
ax.set_xticks(newx_pos_values)
ax.legend()
ax.grid(b=True)

plt.savefig(Star + "_pos"+".jpg")
```

```python
#Grafica de velocidad
xvel_stick = len(Hora)
fig_vel, ax_vel = plt.subplots(figsize=(10,6))
velazim_plot = ax_vel.plot(Hora,vel_az,label="Velocidad Azimut")
velalt_plot = ax_vel.plot(Hora,vel_el,label ="Velocidad elevaci贸n")
ax_vel.set_ylabel("Velocidades 掳/s")
ax_vel.set_xlabel("Tiempo")
xvel_values = ax_vel.get_xticks()
newx_vel_values = [xvel_values[i] for i in range(0,(xvel_stick//2)*2,xvel_stick//10)]
ax_vel.set_xticks(newx_vel_values)
ax_vel.legend()
ax_vel.grid(b=True)
#Astronomic_data_vel.plot()

plt.savefig(Star + "_vel"".jpg")
```
